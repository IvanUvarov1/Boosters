```python
def f(x):
    if x <= -2:
        return 1 - (x+2)**2
    elif x> 2:
        return (x-2)**2 +1
    else: return - (x/2)
```

```python
def modify_list(l):
    i = 0
    while i != len(l):
        if l[i] % 2 == 0:
            l[i] == l[i]//2
            i +=1
        else:
            l.remove(l[i])
    return l
```

### Файловый ввод/вывод

```python
with open('dataset_3363_2.txt') as inf:
    s1 = inf.readline().strip()
stroka = []
c= len(s1)
for i in range(c):
    stroka.append(s1[i])

        
with open('dataset_3363_2.txt', 'w') as ouf:
    for i in range(c-13):
        if stroka[i].isalpha() == True:
            ouf.write(stroka[i], sep ='', end='')
        elif stroka[i+1].isdigit() == True:
            stroka[i] = int(stroka[i])*10 + int(stroka[i+1])
            del(stroka[i+1])
            ouf.write(stroka[i-1]*(int(stroka[i]-1)), sep ='', end='')
        else: ouf.write(stroka[i-1] * (int(stroka[i])-1), sep='', end='')
```

```python
d = {}

f = open('dataset_3363_3.txt')

for line in f.readlines():
    s2 = line.lower().split()
    for i in range(len(s2)):
        if s2[i] not in d.keys():
            d[s2[i]] = 1
        else: d[s2[i]] += 1

maxim = ''        
f = open('dataset_3363_2.txt', 'w')
for key,value in d.items():
    if value == max(d.values()):
        if str(d[(key)]) > maxim:
            maxim = key
            maximum = value
f.write(str(maxim) + ' ' + str(maximum))
f.close()
```

### Задачи по материалам недели 

```python
n = int(input())
a = []
for i in range(n):
    a.append(input().split(';'))

def make_spisok(a):
    spisok = [a]
    spisok.append([0 for i in range(0,5)])
    return spisok

spisok_komand = []
for i in range(n):
    if a[i][0] not in spisok_komand:
        spisok_komand.append(a[i][0])
for i in range(n):
    if a[i][2] not in spisok_komand:
        spisok_komand.append(a[i][2])

dlina_spiska = len(spisok_komand)
tablica = [] 

for i in spisok_komand:
    tablica.append(make_spisok(i))
for i in range(n):
    if a[i][1]>a[i][3]: # Если выиграла домашняя команда
        for j in range(dlina_spiska): # проходим по количеству команд
            if a[i][0] == tablica[j][0]: #Если Зенит = Зенит
                tablica[j][1][1]+=1  # увечличиваем количество выигрышей, третий индекс - результат
            if a[i][2] == tablica[j][0]:
                tablica[j][1][3]+=1 #увеличиваем количество поражений
    if a[i][1]<a[i][3]: # Если выиграла гостевая команда
        for j in range(dlina_spiska):
            if a[i][2] == tablica[j][0]:
                tablica[j][1][1]+=1
            if a[i][0] == tablica[j][0]:
                tablica[j][1][3]+=1
    if a[i][1]==a[i][3]: 
        for j in range(dlina_spiska):            
            if a[i][2] == tablica[j][0]:
                tablica[j][1][2]+=1
            if a[i][0] == tablica[j][0]:
                tablica[j][1][2]+=1
                
for i in range(dlina_spiska): # считаем количество очков и матчей
    tablica[i][1][4] = tablica[i][1][1] * 3 + tablica[i][1][2]
    tablica[i][1][0] = tablica[i][1][1] + tablica[i][1][2] + tablica[i][1][3]

for i in range(dlina_spiska):
    print (str(tablica[i][0])+':'+str(tablica[i][1][0])+' '+str(tablica[i][1][1])+' '+str(tablica[i][1][2])+' '+str(tablica[i][1][3])+' '+str(tablica[i][1][4]))
```
